## DeepIV

- $y$ is structually determined by $p,x,e$:

  $y = g(p,x)+e$
  
  where
  - $p$: treatment
  - $x$: other covariates
  - $e$: error

- Counterfactual prediciton function:

  $h(p,e) = E[y|p,x] = g(p,x)+E[e|x]$
  
  assuming distribution of $e$ is constant as $p$ changes

- Let $z$ be an instrumental variable (IV):

$\begin{align*}
E[y|x,z] &= E[g(p,x)|x,z]+E[e|x,z]\\
         &= \int h(p,x) dF(p|x,z)
\end{align*}$


- DeepIV procedures:
  - stage one: estimate the density $\hat{F}(p|x,z)$
  - stage two: solve $\min_{\hat{h}\in H} \sum_i (y - \int \hat{h}(p,x) dF(p|x,z))$
