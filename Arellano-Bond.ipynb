## Arellano-Bond Estimator

### Linear regression function:

$Y_{i,t} = \beta_0 + \beta_1 * Y_{i,t-1} + \beta_2 * T_{i,t} + \beta_3 * T_{i,t}^2 + \beta_4 * P_{i,t} + \beta_5 * P_{i,t}^2 + \epsilon_{i,t}$

where

- $Y$: growth rate of GDP per capita
- $T$: temperature
- $P$: precipitation
